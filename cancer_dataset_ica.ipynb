```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer_dataset=load_breast_cancer()
```

```python
cancer_dataset.keys()
```

```python
df=pd.DataFrame(cancer_dataset['data'],columns=cancer_dataset['feature_names'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data=scaler.transform(df)
```

```python
from sklearn.decomposition import FastICA
```

```python
ica=FastICA(n_components=2)
```

```python
data_ica=ica.fit_transform(scaled_data)
```

```python
data_ica
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data_ica[:,0],data_ica[:,1],c=cancer_dataset['target'],cmap='coolwarm')
plt.xlabel('First IC')
plt.ylabel('Second IC')
```

```python
##here we can see that the two independent components are able to differnetiate clearly the ,elignin and benignin and we do not see any considerable overlap so we can say that ica is bale to work properly here
```
